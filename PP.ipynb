```python
for i in sent:
    doc = nlp(i)
    #displacy.render(doc, style="dep")

    for token in doc:
        print(f"{token.text:{8}} {token.tag_:{10}} {token.dep_:{10}}   {token.head.text:{9}} {token.head.tag_:{9}} {token.dep_:{10}} {spacy.explain(token.dep_)} ")
    print('___________________________')
```

```python
def PP(sent):
    parse = nlp(sent)
    sim_sent = sent
    hidden =[]
    sub_tree = []
    noun_tag = ['NN', 'NNS', "NNP", "NNPS"]
    
    # find PP
    for word in parse:
        if word.dep_ == 'prep':
            if word.head.tag_ in noun_tag and word.text != 'of':
                sub_tree.append((word.head, word.subtree))
            else:
                sim_sent = sent
    
    # PP subtree into string
    for i in sub_tree:
        w = []
        for word in i[1]:
            w.append(word.text)
        hidden.append((i[0], ' '.join(w)))
    
    # remove pp from sentence
    for i in hidden:
        sim_sent = sent.replace(i[1],'')
    
    # remove uneccessary space due to previous removal
    sim_sent = sim_sent.replace('  ',' ')
    sim_sent = sim_sent.replace(' .','.')
    sim_sent = sim_sent.replace(' :',':')
    sim_sent = sim_sent.replace(' ;',';')
    sim_sent = sim_sent.replace(' ,',',')
    #print(sent)
    return(sim_sent, hidden)
```

```python
for i in sent:
    print(PP(i))
```
